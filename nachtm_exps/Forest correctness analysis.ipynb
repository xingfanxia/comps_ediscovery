```python
import sys
sys.path.append('..')
from lib import *
import pandas as pd
import numpy as np
```

# Read in dataframe

```python
lsa_np = np.load('../data/parsed/lsa_output.npy')
```

```python
metadata = pd.read_pickle('../data/parsed/pickles/pickled_data_test.pickle')
metadata = metadata.loc[metadata['Scenario'] == '401']
metadata = metadata.reset_index(drop=True)
```

```python
lsa_df = pd.DataFrame(lsa_np)
```

```python
df = pd.concat([metadata, lsa_df], axis=1, join_axes=[metadata.index])
df = df.loc[df['Label'] != '-1']
df = df.reset_index(drop=True)
```

```python
cat_features = ['To','From']
features = list(range(100))
features.extend(cat_features + ['Date'])

df = df[features + ['Label', 'ID']]
```

```python
test_df = df[100:].reset_index(drop=True)
```

# Load a forest

```python
forest = RNF(None, None, None, None, None, None, None)
forest.load_rnf('scen1.first100.64trees.5depth.42rs.11maxfeats.100maxin.pickle')
```

# Prediction and analysis

```python
results = forest.predict(test_df)
```

```python
results[1]
```

```python
evalMetrics.evalStats(results[1], test_df)
```

```python
print("Incorrect:")
for i in range(len(results[1])):
    if results[1][i] != test_df["Label"][i]:
        print('Row: {}, confidence: {}'.format(i,max(results[0][i])))
print("Correct:")
for i in range(len(results[1])):
    if results[1][i] == test_df["Label"][i]:
        print('Row: {}, confidence: {}'.format(i,max(results[0][i])))
```

```python
yesIDs = []
noIDs = []
print("Rel and identified rel:")
for i in range(len(results[1])):
    if results[1][i] == '1' and test_df["Label"][i] == '1':
        print(test_df["ID"][i])
        yesIDs.append(test_df["ID"][i])
print("Rel and identified irrel:")
for i in range(len(results[1])):
    if results[1][i] == '0' and test_df["Label"][i] == '1':
        print(test_df["ID"][i])
        noIDs.append(test_df["ID"][i])
```

```python
for doc in metadata[metadata["ID"].isin(noIDs)].iterrows():
    if doc[1]["Scenario"] == '401':
        print(doc[1]["Message-Contents"])
        print("End of message. DAGHKDGHSABGDVF32tgak")
        print('-------------------------------------')
        print('-------------------------------------')        
        print('-------------------------------------')        
```

```python
false_pos = 0
true_pos = 0
false_neg = 0

for doctuple in metadata[600:].iterrows():
    doc = doctuple[1]
    if doc["Scenario"] == '401' and 'enrononline' in doc['Message-Contents'].lower():
        #mark as rel
        if doc["Label"] == '0':
            false_pos += 1
        elif doc["Label"] == '1':
            true_pos += 1
    elif doc["Label"] == '1':
        false_neg += 1

print("FP: {}".format(false_pos))
print("TP: {}".format(true_pos))
print("FN: {}".format(false_neg))
print("precision: {}".format((true_pos)/(true_pos+false_pos)))
print("recall: {}".format((true_pos)/(true_pos+false_neg)))
```
